```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from collections import defaultdict
```

```python
data = np.load('others/others/logs/static_real_07232020-08-33-23//diagnostics.npy', allow_pickle=True)
```

```python
def process_data(data):
    new_dict = defaultdict(list)
    for t, d in enumerate(data):
        for k, val in d.items():
            assert(len(new_dict[k]) == t)
            if val is None:
                val = 0
            if isinstance(val, list):
                val = val[0]
            new_dict[k].append(val)
            
        new_dict['epoch'].append(t)
        new_dict['training_success_rate'].append(d['num_success']/100)
    return new_dict
```

```python
bigimage =  process_data(np.load('../mobilemanipulation/others/others/logs/static_real_07232020-08-33-23/diagnostics.npy', allow_pickle=True))

key = 'training_successes'
plt.plot(bigimage['num_samples_total'], bigimage[key], label='real')

plt.legend()
plt.title(key)
plt.show()
key = 'eval_success_ratio'
plt.plot(bigimage['num_samples_total'], bigimage[key], label='real')

plt.legend()
plt.title(key)
plt.show()
```
